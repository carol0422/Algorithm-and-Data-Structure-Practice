```python
import random 
import numpy as np
from time import perf_counter_ns
from matplotlib import pyplot as plt
```

# Insertion Sorting function 

```python
def insertion_practice(input_seq):   
   
    counter=1
    while counter<len(input_seq):  
        key = input_seq[counter]
        counter_2 =counter-1 
        while (counter_2>=0 and input_seq[counter_2]>key):
            input_seq[counter_2+1] = input_seq[counter_2]
            counter_2 = counter_2-1   
        input_seq[counter_2+1] = key
        counter=counter+1    
    return input_seq
```

### Random List Sorting:  Printing the time taken when sorting a randomized sorted list of length ranging from 2 power of 1 to 2 to the power of 18

```python

total_time = []
y_axis = []


counter= 1
while counter <18:
    random_list = list(range(2**counter))
    y_axis.append(2**counter)
    
    random.shuffle(random_list)
    print("===============Start Loop================")
    print("length of the list ::" + str(len(random_list)))
    
    start_time = perf_counter_ns()
    temp = insertion_practice(random_list)
    total_time.append(perf_counter_ns() - start_time)
    print("Time taken :: " + str(perf_counter_ns() - start_time))
    print("===============End Loop================")
    counter+=1
```

## Worst, average, and best case Scenario:

Based on this criteria, one could apply the O notation to evaluate the performance of the algorithm at different ranges.  For a list whose length is n, under the best and worst cases the number of comparisons and moves required in the insertion sorting algorithm are shown in the following table.


| Criteria | Best | Worst | 
| --- | --- | --- |
| comparison | n-1 | 2+3+4+...+n = (n*n+n-2)/2 |
| move | 0 | 1+2+3+...+(n-1) = (n*n)/2 | 

 
Because elements do not move after every comparison, the number of moves is less than the number of comparisons. Therefore, we take the comparison times as an index to measure the time complexity of the algorithm. The time complexity of insertion sort is shown below.



| Scenario | Time complexity | 
| --- | --- |
| Best | O(n) | 
| worst | O(n*n) | 
| average | O(n*n) |
