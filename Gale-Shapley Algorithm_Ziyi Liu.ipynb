```python
# Packages
import random
import math
from time import perf_counter_ns
from matplotlib import pyplot as plt
import numpy


class GSA: # Gale-Shapley Algorithm
    def __init__(self, count): 
        self.n = count

    def define_groups(self, group1, group2):
        self.M_F = [group1, group2]

    def get_preference_grp1(self, grp1):
        self.M = grp1 # grp1=elements in group 1

    def get_preference_grp2(self, grp2):
        self.F = grp2 # grp2=elements in group 2

    def select_proposing_group(self, group_index):  # 0=first 1=second
        if group_index == 0:
            self.proposing_group = 0
            self.proposed_group = 1
        else:
            self.proposing_group = 1
            self.proposed_group = 0
            

    def create_unmatched_list(self):  # list of unmatched people from proposing group
        self.unmatched = []
        for i in self.M_F[self.proposing_group]:
            self.unmatched.append(i)

    def create_current_partner_matrix(self):  # pairs of matched partners
        self.current_partners = [[], []]
        for i in range(self.n):
            self.current_partners[0].append('_')
            self.current_partners[1].append('_')

    def create_proposed_matrix(self):  # matrix to mark the proposed people
        self.proposed = []
        for i in range(self.n):
            var = []
            for j in range(self.n):
                var.append(0)
            self.proposed.append(var)

    def calculate_gsa(self):  
        while (len(self.unmatched) > 0):
            current_male = self.unmatched.pop(0)
            i = self.M_F[self.proposing_group].index(current_male)
            for j in range(self.n):
                if self.proposed[i][j] == 0:  # this male hasnt proposed the female at pos j
                    self.proposed[i][j] = 1
                    current_female = self.M[i][j]
                    male_index = self.M_F[self.proposing_group].index(current_male)
                    female_index = self.M_F[self.proposed_group].index(current_female)
                    if current_female in self.current_partners[
                        self.proposed_group]:  # the female is already engaged and needs to check priority
                        defending_male = self.current_partners[self.proposing_group][
                            self.current_partners[self.proposed_group].index(current_female)]  # rival of proposing male
                        defending_male_pos = self.F[female_index].index(defending_male)
                        current_male_pos = self.F[female_index].index(current_male)
                        if current_male_pos < defending_male_pos:  # the proposing male has higher priority than defending male in view of the female
                            self.current_partners[self.proposing_group][
                                self.M_F[self.proposing_group].index(defending_male)] = '_'
                            self.current_partners[self.proposed_group][
                                self.M_F[self.proposing_group].index(defending_male)] = '_'
                            self.current_partners[self.proposing_group][
                                self.M_F[self.proposing_group].index(current_male)] = current_male
                            self.current_partners[self.proposed_group][
                                self.M_F[self.proposing_group].index(current_male)] = current_female
                            self.unmatched.insert(self.proposing_group, defending_male)
                            break
                    else:  # engage the female who is not engaged to the male
                        self.current_partners[self.proposing_group][male_index] = current_male
                        self.current_partners[self.proposed_group][male_index] = current_female
                        break
        return self.current_partners

sm=GSA(5)
sm.define_groups(['A','B','C','D','E'],['L','M','N','O','P'])
sm.get_preference_grp1([['O','M','N','L','P'],['P','N','M','L','O'],['M','P','L','O','N'],['P','M','O','N','L'],['O','L','M','N','P']])
sm.get_preference_grp2([['D','B','E','C','A'],['B','A','D','C','E'],['A','C','E','D','B'],['D','A','C','B','E'],['B','E','A','C','D']])
sm.select_proposing_group(0)
sm.create_unmatched_list()
sm.create_current_partner_matrix()
sm.create_proposed_matrix()
print(sm.calculate_gsa())


#I tried to measure the time complexity but failed...I'll retry it this weekend.
#creating a list of 10000
M = []
for i in range(10000):
    M.append(i)
    
F = []
for i in range(10000):
    F.append(i)
    
    
list_10000a = []
for i in range(10000):
    list_10000a.append(i)
random.shuffle(list_10000a)

list_10000b = []
for i in range(10000,20000):
    list_10000b.append(i)
random.shuffle(list_10000b)

#measure the time of matching lists of 10000
time = []
sm=GSA(10000)
sm.define_groups(M,F)

a = numpy.array(None)
shuffle_resulta = numpy.array(list_10000a)
for i in(0,10000):
    a =numpy.r_[a,shuffle_resulta]


b = numpy.array(None)
shuffle_resultb = numpy.array(list_10000b)
j = 0
for j in(0,10000):
    b =numpy.r_[b,shuffle_resultb]

sm.get_preference_grp1(b)                  
sm.get_preference_grp2(a)
start_time = perf_counter_ns()
sm.select_proposing_group(0)
sm.create_unmatched_list()
sm.create_current_partner_matrix()
sm.create_proposed_matrix()
result = sm.calculate_gsa()
time.append(perf_counter_ns() - start_time)
```
